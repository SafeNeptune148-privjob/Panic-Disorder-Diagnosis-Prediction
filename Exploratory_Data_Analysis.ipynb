```python
#Importación de las librerías correspondientes así como de la configuración del data set
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
pd.set_option("display.max_columns",None)
```

```python
#Creación del dataset con datos de entrenamiento
df = pd.read_csv("panic_disorder_dataset_training.csv")
df.head(15)
```

```python
#Buscamos que los valores tengan coherencia, ver si hay desbalence etc.
df.describe()
```

```python
#Eliminamos el ID 
df = df.drop(["Participant ID"], axis=1)
```

```python
df
```

```python
#Verificación de datos nulos
df.isnull().sum()
```

```python
#Llenamos los valores nulos de manera aleatoria, para que haya más variedad
nan_col = ["Medical History","Psychiatric History","Substance Use"]
for col in nan_col:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
df["Medical History"].value_counts().plot(kind="barh")
plt.show()
```

```python
#Verificación de que los datos nulos se hayan llenado de manera correcta
df.isnull().sum()
```

```python
#Visualización de los tipos de datos
df.dtypes
```

```python

pd.crosstab(df["Gender"],df["Age"]).plot(kind = "bar", stacked=True, figsize = (20,10))
plt.show()
```

```python
#Visualizammos que tan balanceados están los pacientes con el desorden de ataque de pánico
df["Panic Disorder Diagnosis"].value_counts().plot(kind="bar")
plt.show()
```

Como se puede ver está muy desbalanceado, de modo que esto que nos puede jugar en contra al momento de entrenar los modelos, pues al haber una gran falta de diágnosticos positivos, no aprenderá bien a clasificar estos datos.

```python
df
```

```python
#Asociación entre las variables
from scipy.stats import chi2_contingency
for col in df.columns.drop(["Panic Disorder Diagnosis"]):
    contingency_table = pd.crosstab(df[col],df["Panic Disorder Diagnosis"])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print("Chi-square Statistics: {}, p-value: {} for the feature: {}".format(chi2,p,col))
    
```

Aquí se hace la prueba chi cuadrada para ver la relevancia de los datos.

```python
df
```

```python
plt.figure(figsize=(10,6))
sns.stripplot(x = df["Age"],y = df["Demographics"], jitter=True, alpha=0.3)
plt.xlabel("Age")
plt.ylabel("Panic Disorder Diagnosis")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
df_sample,_ = train_test_split(df, train_size=10000, stratify=df["Panic Disorder Diagnosis"], random_state=42)
df_sample["Panic Disorder Diagnosis"].value_counts()
```

Aquí como se puede ver con una muestra de 10000 datos, solo el 5% tiene diágnosticado el desorden de ataques de pánico

## Visualización de las diferentes variables en gráficas de barras

## Gráfica de factores de vida

```python
import seaborn as sns
import matplotlib.pyplot as plt

colors = {0:"#003F7A",1:"#000000"}
plt.figure(figsize=(8,5))
ax = sns.histplot(data=df_sample, x="Lifestyle Factors", hue="Panic Disorder Diagnosis", 
             stat="percent", multiple="fill", shrink=0.8, palette = colors)


plt.xlabel("Lifestyle Factors")
plt.ylabel("Percentage")
plt.title("Panic Disorder Distribution by Lifestyle Factors (Sampled Data)")
plt.show()
```

## Gráfica de edades 

```python
plt.figure(figsize=(8,5))
ax = sns.histplot(data=df_sample, x="Age", hue="Panic Disorder Diagnosis", 
             stat="percent", bins=10,multiple="fill", shrink=0.8, palette = colors)


plt.xlabel("Age")
plt.ylabel("Percentage")
plt.title("Panic Disorder Distribution by Age (Sampled Data)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
ax = sns.histplot(data=df_sample, x="Current Stressors", hue="Panic Disorder Diagnosis", 
             stat="percent", multiple="fill", shrink=0.8, palette = colors)


plt.xlabel("Current Stressors")
plt.ylabel("Percentage")
plt.title("Panic Disorder Distribution by Current Stressors (Sampled Data)")
plt.show()
```
